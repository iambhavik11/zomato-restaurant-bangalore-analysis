```python
filter_df = df.dropna()
filter_df = filter_df.reset_index(drop=True)
filter_df.rename(columns={'approx_cost(for two people)':'cost','listed_in(type)':'in_type'},inplace=True)
filter_df = filter_df[['name','online_order','book_table','votes','location','rest_type','dish_liked','cuisines','cost','in_type']]
filter_df
```

## **Exploratory Analysis and Visualization**

#### Most Liked Dish

```python
liked_dish = filter_df[(filter_df['location']=='Whitefield') & (filter_df['dish_liked'])]
liked_dish[['name','cost']].sort_values('cost').tail()
```

```python
ldish = liked_dish['dish_liked'].value_counts().head()
ldish['dishes'] = ldish.keys
ldish['counts'] = ldish.values
```
